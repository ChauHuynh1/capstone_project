## Analysing

```python
import cv2
import numpy as np

# Load the thermal image
image_path = "data/Thermal/snap_1_ (1).jpg"
thermal_image = cv2.imread(image_path)

# Preprocess the image (if needed)
# For thermal images, you may need to apply appropriate image processing techniques like adjusting brightness, contrast, or thresholding to enhance cell visibility.

# Convert the image to grayscale
gray_image = cv2.cvtColor(thermal_image, cv2.COLOR_BGR2GRAY)

# Apply edge detection to identify cell boundaries
edges = cv2.Canny(gray_image, 50, 150)

# Find contours in the image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Define a list to store the cropped cell images
cropped_images = []

# Iterate over the contours and extract the individual cells
for contour in contours:
    # Find the bounding rectangle around the contour
    x, y, w, h = cv2.boundingRect(contour)
    
    # Crop the cell region from the thermal image
    cell_image = thermal_image[y:y+h, x:x+w]
    
    # Append the cropped cell image to the list
    cropped_images.append(cell_image)

# Save the cropped images
for i, cropped_image in enumerate(cropped_images):
    cv2.imwrite(f'cell_{i+1}.jpg', cropped_image)

print("Cell cropping completed!")
```
